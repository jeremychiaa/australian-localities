```
# Create function to do the same for NSW and QLD
def save_csv(state):
  df1 = df.loc[df.state == state]
  df1 = df1[['postcode', 'locality', 'region']]
  r1 = df1.loc[df1.region == 'R1']
  r2 = df1.loc[df1.region == 'R2']
  r3 = df1.loc[df1.region == 'R3']
  pd.DataFrame.to_csv(r1, f'{state}_r1.csv')
  pd.DataFrame.to_csv(r2, f'{state}_r2.csv')
  pd.DataFrame.to_csv(r3, f'{state}_r3.csv')
```

```
states = ['NSW', 'QLD']
```

```
for state in states:
  save_csv(state)
```
